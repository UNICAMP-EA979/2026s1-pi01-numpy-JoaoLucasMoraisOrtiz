### Exercício 2.4



Crie uma função que tenha como entradas:
- Inteiro **m** que corresponde às linhas de uma matriz;
- Inteiro **n** que corresponde às colunas de uma matriz;

e que retorne:
- Uma matriz **X** de shape (m, n), em que cada elemento X[i,j] seja a distância Euclidiana entre a posição [i,j] e a posição [0,0] (ou seja, X[i,j] = módulo de (i,j)).

Teste a função com os valores **m=5** e **n=4**.

<details>
  <summary>Dicas</summary>

  Você pode pensar nessa questão em duas partes: gerar os índices + calcular o módulo deles. Cada parte pode ser feita de várias formas:

  Gerar os índices
  - [numpy.indices](https://numpy.org/doc/stable/reference/generated/numpy.indices.html)
  - [numpy.meshgrid](https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html) e [numpy.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html)
  - [numpy.argwhere](https://numpy.org/doc/stable/reference/generated/numpy.argwhere.html) e [numpy.ones](https://numpy.org/doc/stable/reference/generated/numpy.ones.html)

  A função [numpy.ndarray.transpose](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.transpose.html) pode ser útil.

  Calcular o módulo
  - [numpy.linalg.norm](https://numpy.org/doc/stable/reference/generated/numpy.linalg.norm.html)
  - [numpy.power](https://numpy.org/doc/stable/reference/generated/numpy.power.html) e [numpy.sqrt](https://numpy.org/doc/stable/reference/generated/numpy.sqrt.html)
  
  
</details>

```python
# Seu código aqui

def plano_distancias(m, n):

    # precisamos pegar dois eixos
    I, J = np.indices((m, n))

    #agora é só percorrer esses dois eixos, calculando
    return np.sqrt((I - m//2)**2 + (J - n//2)**2)

plano_distancias(5, 4)
```

## Parte 5 - Geração automática de imagens sintéticas e manipulação de arquivos (opcional para EA979) 🤖🚫

### Exercício 5.1

**a)** Quando geramos imagens sintéticas, estamos interessados em criar um conjunto de imagens com algumas características pré-definidas, mas com certa variabilidade. Para exemplificar esse processo, vamos automatizar a criação dos quadrados do exercício 3.1.

Crie uma função que tenha como entradas:
- Inteiro **n** que corresponde ao número de imagens a serem criadas;
- Inteiro **h** que corresponde à "altura" de cada imagem (com "largura" 2\***h**);

e que retorne:
- Um array **I** de shape (n, h, 2\*h, 3), em que cada **I**[n] corresponda a uma imagem similar à imagem gerada no exercício 3.1.b), mas colorida.

A ideia é que a função crie um conjunto de **n** imagens, todas com os "quadrados menores" aproximadamente centralizados nos "quadrados maiores". O tamanho dos "quadrados menores" deve ser igual para todas as imagens do conjunto e definido de forma aleatória. As duas cores utilizadas devem variar para cada imagem e também são definidas aleatoriamente.

Você deve definir certos limites para garantir que a geração das imagens faça sentido: qual a altura mínima das imagens? qual o tamanho mínimo dos quadrados menores? qual a distância mínima entre as bordas dos quadrados menor/maior (para que um não se sobreponha ao outro)?

Teste a função com diferentes valores de entrada, tentando sempre plotar as imagens de cada conjunto gerado em um único plot. Comente sobre as decisões tomadas na criação da função (como os limites que foram definidos) e seu comportamento esperado (como a centralização dos quadrados menores em relação às dimensões da imagem).

```python
# Seu código aqui
def gerar_conjunto_quadrados(n, h):
    # Limites mínimos para manter geometria válida
    if n <= 0:
        raise ValueError("n deve ser > 0.")
    if h < 20:
        raise ValueError("h deve ser >= 20 para caberem os quadrados com margem.")

    # Distância mínima entre bordas (quadrado menor vs quadrado maior)
    margem = h // 12

    # Tamanho do quadrado menor: único para todo o conjunto, sorteado uma vez
    tam_min = max(4, h // 6)
    tam_max = h - 2 * margem
    if tam_min > tam_max:
        raise ValueError("Não foi possível definir tamanho do quadrado menor com os limites escolhidos.")
    s = np.random.randint(tam_min, tam_max + 1)

    # divisão arredondada para ficar "aproximadamente" centralizado
    jitter_max = max(0, h // 20)

    I = np.zeros((n, h, 2 * h, 3), dtype=np.uint8)

    for k in range(n):
        # Duas cores aleatórias (RGB), uma para cada quadrado maior
        cor_esq = np.random.randint(0, 256, size=3, dtype=np.uint8)
        cor_dir = np.random.randint(0, 256, size=3, dtype=np.uint8)

        # Garante contraste mínimo entre as duas cores
        while np.linalg.norm(cor_esq.astype(int) - cor_dir.astype(int)) < 60:
            cor_dir = np.random.randint(0, 256, size=3, dtype=np.uint8)

        # Metades da imagem: dois quadrados maiores
        I[k, :, :h, :] = cor_esq
        I[k, :, h:, :] = cor_dir

        # Centro nominal dos quadrados menores em cada metade + pequeno jitter
        dr = np.random.randint(-jitter_max, jitter_max + 1)
        dc = np.random.randint(-jitter_max, jitter_max + 1)

        r0 = np.clip(h // 2 - s // 2 + dr, margem, h - margem - s)
        c0_esq = np.clip(h // 2 - s // 2 + dc, margem, h - margem - s)
        c0_dir = np.clip(h + h // 2 - s // 2 + dc, h + margem, 2 * h - margem - s)

        # Quadrado menor com cor invertida em cada metade
        I[k, r0:r0 + s, c0_esq:c0_esq + s, :] = cor_dir
        I[k, r0:r0 + s, c0_dir:c0_dir + s, :] = cor_esq

    return I


def plotar_conjunto(I, titulo="Conjunto de imagens"):
    n = I.shape[0]
    cols = min(n, 5)
    rows = int(np.ceil(n / cols))

    plt.figure(figsize=(3 * cols, 3 * rows))
    for i in range(n):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(I[i])
        plt.axis("off")
        plt.title(f"I[{i}]")
    plt.suptitle(titulo, y=1.02)
    plt.tight_layout()
    plt.show()


# Testes com diferentes entradas
A = gerar_conjunto_quadrados(n=4, h=40)
B = gerar_conjunto_quadrados(n=6, h=60)
C = gerar_conjunto_quadrados(n=8, h=90)

print("A:", A.shape, "B:", B.shape, "C:", C.shape)
print("Limites adotados: h>=20, margem mínima=max(2, h//12), tamanho mínimo do quadrado menor=max(4, h//6).")

plotar_conjunto(A, "Conjunto A (n=4, h=40)")
plotar_conjunto(B, "Conjunto B (n=6, h=60)")
plotar_conjunto(C, "Conjunto C (n=8, h=90)")
```

Existem diversas funções que calculam dados [matemáticos](https://numpy.org/doc/stable/reference/routines.math.html) e [estatísticos](https://numpy.org/doc/stable/reference/routines.statistics.html) de um array. Esses dados podem ser muito úteis para diversas aplicações em imagens.

**a)** Observe o código abaixo. O que ele está realizando? Qual a importância do valor $-0.397$? Este código funcionaria com outra imagem? Na sua explicação, utilize alguma forma de visualizar o que está acontecendo.

```python
url = "https://raw.githubusercontent.com/UNICAMP-EA979/EA979-Recursos/b30cf8a41c9b04a7aa20a47e668cee2debd52bc5/imgs/keyb.tif"
img = np.array(read_image_url(url))

#primeiro fui visualizar a imagem para entender o que ela era
print(img.shape)
plt.figure(figsize=(4, 4))
plt.imshow(img, cmap='gray')
plt.title('Imagem original')
plt.axis('off')

# descobri que a imagem é uma calculadora em grayscale.

#estamos aqui tentando pegar algumas métricas sobre a matriz (imagem)
ops = [np.mean, np.max, np.min, np.std, np.median]
features = []
for op in ops:
    # para cada uma das operações que setamos anteriormente,
    # vamos aplicala para o eixo X (linhas), mas vamos dividir por 255, que é o valor máximo que pode haver
    # assim vamos ter valores entre 0 e 1.
    f0 = op(img, axis=0)/255

    # e para o eixo Y (colunas) - mesma coisa para normalizar os valores (dividir por 255)
    f1 = op(img, axis=1)/255

    # então fazemos o append para guardar essas informações, após fazer a multiplicação entre os dois eixos (x, y)
    features.append(np.outer(f1, f0))

# eu não entendi agora porque que fizemos todo o processo de pegar várias métricas,
# sendo que usamos só duas: a média e o máximo.
# mas em fim, basicamente temos a remoção da média, e ficamos então com a diferença entre o máximo e a média, que é o que tem mais contraste
test = (features[0]-features[1])/2 # o 2 é para normalizar.

print(np.mean(test))
# o resultado é um array onde os valores mais negativos correspondem a regiões onde a média é maior que o máximo, ou seja, regiões mais escuras. Já os valores mais positivos correspondem a regiões onde o máximo é maior que a média, ou seja, regiões mais claras.
# então, ao criar a mascara test < -0.397, estamos selecionando as regiões onde a média é significativamente maior que o máximo, ou seja, as regiões mais escuras da imagem. O valor -0.397 é um limiar escolhido para destacar essas regiões escuras.
# -0.397 é interessante pois é menor que a média dos valores do array test (-0.260), então ele vai pegar as regiões que são mais escuras do que a média geral, que são as teclas.
mask = test < -0.397

plt.figure(figsize=(4, 4))
plt.imshow(mask, cmap='gray')
```

**b)** Você consegue alterar o código para que ele realiza a mesma funcionalidade, mas de forma mais eficiente? Faça uma versão que leia cada elemento da imagem apenas 2 vezes (exemplo: ler uma vez cada linha e cada coluna). Além da eficiência, este método tem alguma outra vantagem? Comente sobre.

```python
# Versão mais eficiente: usa apenas média e máximo (o que realmente entra no cálculo final)
url = "https://raw.githubusercontent.com/UNICAMP-EA979/EA979-Recursos/b30cf8a41c9b04a7aa20a47e668cee2debd52bc5/imgs/keyb.tif"
img = np.array(read_image_url(url))

# Normaliza uma única vez
imgf = img.astype(np.float32) / 255.0

# Lê cada pixel no eixo das linhas e no eixo das colunas (uma vez por eixo)
row_mean = imgf.mean(axis=1)
col_mean = imgf.mean(axis=0)
row_max = imgf.max(axis=1)
col_max = imgf.max(axis=0)

# Reconstrói os mapas 2D com produto externo
feat_mean = np.outer(row_mean, col_mean)
feat_max = np.outer(row_max, col_max)

# Mesma funcionalidade lógica do item (a): contraste entre média e máximo
test = (feat_mean - feat_max) / 2.0
mask = test < -0.397

# Visualização
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagem original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(test)
plt.title('Mapa test')
plt.axis('off')
plt.colorbar(fraction=0.046, pad=0.04)

plt.subplot(1, 3, 3)
plt.imshow(mask, cmap='gray')
plt.title('Máscara final')
plt.axis('off')

plt.tight_layout()
plt.show()

print('Comentário:')
print('- Eficiência: evitamos calcular min, std e mediana, pois não eram usadas no resultado final.')
print('- Menos leituras úteis da imagem: focamos apenas no que entra em test (média e máximo por linha/coluna).')
print('- Outra vantagem: código mais simples, mais fácil de depurar e de ajustar para outras imagens/limiares.')
```
